```python
def randomSearch():
  characters = "abcdefghijklmnopqrstuvwxyz0123456789"
  randomCharacter = characters[random.randint(0, len(characters) - 1)]
  randomNum = random.randint(0, 2)

  if randomNum == 0:
    randomSearchStr = randomCharacter + '%'
  elif randomNum == 1:
    randomSearchStr = '%' + randomCharacter
  elif randomNum == 2:
    randomSearchStr = '%' + randomCharacter + '%'

  return randomSearchStr
```

```python

for x in range (0, 10):

  songOffset = random.randint(0, 1000)
  s = randomSearch()

  track = sp.search(q=s, type="track", limit=1, offset=songOffset)
  track_uri = track['tracks']['items'][0]['preview_url']
  track_name = track['tracks']['items'][0]['name']

  url = requests.get(track_uri)
  filename = track_name + '.mp3'

  with open(filename, 'wb') as f:
      f.write(url.content)
```
