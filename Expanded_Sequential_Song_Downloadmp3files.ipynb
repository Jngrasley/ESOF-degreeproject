```python
# Random Search function to provide text inputs for song searches
def randomSearch():
  # define list of possible characters
  characters = "abcdefghijklmnopqrstuvwxyz0123456789"
  # from this list, take 1 character randomly
  randomCharacter = characters[random.randint(0, len(characters) - 1)]

  #generate a random number between 0-2
  randomNum = random.randint(0, 2)
  # generate a random number to determine the formatting of the input text within the search
  if randomNum == 0:
    # declares that the input can appear only before other text in the search input
    randomSearchStr = randomCharacter + '%'
  elif randomNum == 1:
    # declares that the input can appear only after other text in the search input
    randomSearchStr = '%' + randomCharacter
  elif randomNum == 2:
    # declares that the input can appear anywhere within the search input (i.e. before, after, or in the middle)
    randomSearchStr = '%' + randomCharacter + '%'
  
  return randomSearchStr
```
